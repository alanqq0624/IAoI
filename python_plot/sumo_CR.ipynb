```python
import numpy as np
import csv

import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter  # useful for `logit` scale
from matplotlib.ticker import FormatStrFormatter
# %matplotlib inline
```

```python
xtick_arr = {
#     'TTC': ['TTC-IAoI', 'TTC-TAoI', 'TTC-Base'], 
    'TTC': ['IAoI', 'TAoI', '10Hz'], 
#     'DRAC': ['DRAC-IAoI', 'DRAC-TAoI', 'DRAC-Base']
    'DRAC': ['IAoI', 'TAoI', '10Hz'],
    'CBR': ['0.2', '0.4', '0.6', '0.8']
}

data = [
#     [3, 6, 33], # simple TTC
#     [14, 20, 27], # simple DRAC 1
#     [2, 4, 10], # simple DRAC 2
#     [108, 153, 295], # Lust TTC
#     [102, 134, 238],  # Lust DRAC
    
    [34, 41, 68], # simple TTC NEW
    [14, 20, 27], # simple DRAC 1
    [108, 153, 295], # Lust TTC
    [102, 134, 238],  # Lust DRAC
]

data_cbr = [
#     [80, 79, 75, 79], # LS TTC 
#     [101, 100, 95, 98], # LS DRAC
    [42, 39, 37, 39], # LS TTC NEW
    [18, 17, 15, 16], # LS DRAC NEW
    [103, 103, 101, 105], # Lust TTC NEW
    [105, 98, 96, 100], # Lust DRAC NEW
]

font_size = 30

# labels = ["RTC+", r"$\Delta_x$ = 50ms", r"$\Delta_x$ = 100ms"]
```

```python
def CollisionRate(input_data, data_index, data_size, xtick_arr_index):
    
    x = []
    y = []
    
    x = np.arange(data_size)
    width = 0.5

#     with open('Data_OLD/'+ filename+'_CR_RTC.csv','r') as csvfile:
#         lines = csv.reader(csvfile, delimiter=',')
#         for row in lines:
#             y.append(float(row[1])) 
    
    y = input_data[data_index]
    
    plt.bar(x, y, width, color=['tab:blue', 'tab:orange', 'tab:green', 'tab:purple']) 
    plt.xticks(x , (xtick_arr[xtick_arr_index]))
    
    plt.xlabel('Rate Control Algorithm')
    plt.ylabel('Collision Risk')
#     plt.legend(labels)
#     plt.title(case)
    plt.grid()
```

```python
import matplotlib.pyplot as plt
import csv
from matplotlib.ticker import NullFormatter  # useful for `logit` scale

fig = plt.figure(figsize=(6.5, 5))
plt.rcParams.update({'font.size': 10})
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"]

plt.subplot(221)             
CollisionRate(data, 0, 3, 'TTC')

plt.subplot(222) 
CollisionRate(data, 1, 3,'DRAC')
#plt.legend()

plt.subplot(223) 

CollisionRate(data, 2, 3,'TTC')

plt.subplot(224) 

CollisionRate(data, 3, 3,'DRAC')

plt.gca().yaxis.set_minor_formatter(NullFormatter())

plt.subplots_adjust(top=0.9, bottom=0.1, left=0.10, right=0.95, hspace=0.7,wspace=0.5)
plt.tight_layout()

# plt.savefig("Results_Collision_Rate_Case1.pgf")
plt.savefig("Results_Collision_Rate_Case1.eps")
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
from matplotlib.ticker import NullFormatter  # useful for `logit` scale

fig = plt.figure(figsize=(6.5, 5))
plt.rcParams.update({'font.size': font_size})
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Times New Roman"]

CollisionRate(data, 0, 3, 'TTC')

# plt.gca().yaxis.set_minor_formatter(NullFormatter())
fig.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))

plt.savefig("Results_Collision_Rate_Simple_TTC.eps", format='eps',  bbox_inches='tight')
plt.show()
```
