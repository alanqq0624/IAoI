```python
def BrakeRate(data_index, xtick_arr_index, ylimL, ylimU, ylabelOff):
    
    x = []
    y = []
    
    x = np.arange(len(data[data_index]))
    width = 0.5
    
    y = data[data_index]
    
#     ax.bar(x, y, width, color=['tab:blue', 'tab:orange', 'tab:green']) 
    plt.bar(x, y, width, color=['tab:blue', 'tab:orange', 'tab:green', 'tab:purple']) 
    
    plt.ylim(ylimL, ylimU)
    
    plt.xticks(x , (xtick_arr[xtick_arr_index]))
#     plt.yticks(np.arange(ylimL, ylimU, (ylimU-ylimL)/2 ))
#     plt.subplots_adjust(hspace=0.25)
    
    plt.xlabel('Rate Control Algorithm')
    plt.ylabel('Avg. MBR')
    
#     plt.legend(labels)
#     plt.title(case)
    plt.grid()
```
